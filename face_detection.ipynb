```python
import cv2
import warnings

warnings.filterwarnings('ignore')
```

```python
# Carrega a imagem
imagem = cv2.imread('/home/leticia/Downloads/freepik-export-20240425134308RSyq.jpeg')
```

```python
# Arquivo pré treinado para detectar faces - Específico para frontal.
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
```

```python
# Converte para escala de cinza
gray = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)
```

```python
# Detecta as faces na imagem
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
```

```python
# Desenha os retângulos nas faces detectadas
for (x, y, w, h) in faces:
    cv2.rectangle(imagem, (x, y), (x+w, y+h), (0, 255, 0), 3)
```

```python
# Mostra a imagem com as faces detectadas
cv2.imshow('Faces Detectadas', imagem) #titulo e imagem
cv2.waitKey(0)
cv2.destroyAllWindows()
```
